```python
import numpy as np
import matplotlib.pyplot as plt

# 
x1 = np.linspace(0, 1, 100)
x2 = np.linspace(1, 2, 100)
y = np.linspace(0, 2 * np.pi, 100)
X1, Y1 = np.meshgrid(x1, y)
X2, Y2 = np.meshgrid(x2, y)
Z1 = np.sin(X1) * np.cos(Y1)
Z2 = np.cos(X2) * np.sin(Y2)

# contourf
levels = np.linspace(-1, 1, 100)
cmap = 'viridis'

# 
fig, ax = plt.subplots()

# contourf
contourf1 = ax.contourf(X1, Y1, Z1, levels=levels, cmap=cmap)

# contourf
contourf2 = ax.contourf(X2, Y2, Z2, levels=levels, cmap=cmap)

# x
ax.set_xlim(0, 2)

# contourf
plt.colorbar(contourf1, ax=ax)

plt.show()
```
